#### 인스턴스(멤버) 변수
- 객체마다 따로따로 가지는 데이터
- 'self.변수명' 형태로 작성

```python
# 자동차 클래스 설계
# 속성 : 색상, 속도 => 멤버 변수
# 동작 : 전진,후진,속도 올리기, 속도 내리기...

class Car:
    def __init__(self,color,speed):
        # 인스턴스(멤버) 변수
        self.color = color
        self.speed = speed

    # 인스턴스(멤버) 메소드(함수)
    def upspeed(self,value):
        self.speed = self.speed + value
        print(f"현재 속력 : {self.speed}")

    # 인스턴스(멤버) 메소드(함수)
    def downspeed(self,value):
        self.speed = self.speed - value
        print(f"현재 속력 : {self.speed}")

    # <__main__.Car object at 0x0000029F26A0DF90> => 보고싶은 문자로 변경
    # <__main__.Car object at 0x0000029F26A0DE50> => 보고싶은 문자로 변경
    def __str__(self):
        return f"Car(color={self.color},speed={self.speed})"
```

```python
# 객체 생성
# 변수명 = 클래스명()

# Car.__init__() 자동호출
# 인스턴스 변수 초기화
car1 = Car("Red",50)
car2 = Car("Black",70)

print(car1) # Car(color=Red, speed=50)
# <__main__.Car object at 0x0000029F26A0DF90> => momory address
print(car2) # Car(color=Black, speed=70)
# <__main__.Car object at 0x0000029F26A0DE50> => memory address
```

|-------|--------|-----------|--------|

|'@classmethod'|'cls'|클래스 변수만 | 클래스 단위 동작 |
|'@staticmethod'| 없음 | 없음(매개변수로 받은 것만) | 클래스와 관련은 있지만 독립적인 계산 |

```python
# Account 클래스 작성
# 멤버 변수 : 계좌주(owner), 잔액(balance)
#   메소드 : deposit(money) : 잔액 = 잔액 + 입금액
#   메소드 : withdraw(money) : 잔액 = 잔액 - 출금액
#   메소드 : showBalance(money) : 현재 잔액 출력

class Account:

    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance

    def deposit(self,money):
        self.balance = self.balance + money
        print(f"현재 잔액 : {self.balance}원")

    def withdraw(self,money):
        self.balance = self.balance - money
        print(f"현재 잔액 : {self.balance}원")

    def showBalance(self):
        print(f"현재 잔액 : {self.balance}원")

acc1 = Account("홍길동",10000)

acc1.deposit(5000)
acc1.withdraw(3000)

acc1.showBalance()
```

```python
# Book 클래스 작성
# 클래스 변수 : book_count
# 멤버 변수 : 제목(title), 저자(author)
#   메소드 : showInfo() => 제목 : 파이썬
#                          저자 : 홍길동
# 현재 책 권수 출력

class Book:
    book_count = 0
    def __init__(self,title,author):
        self.title = title
        self.author = author
        Book.book_count += 1
        print(f"현재 책 권수 : {Book.book_count}")

    def showInfo(self):
        print(f"제목 : {self.title}\n저자 : {self.author}")

book1 = Book("파이썬","홍길동")
book1.showInfo()
```
