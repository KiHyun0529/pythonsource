```python
num = 5

# if 문 : bool 자료형
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70이상이면 C, 나머지는 F

score = 90

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
```

```python
a = 12
b = 6
c = 19
# or
# a,b,c = 12,6,19

# 가장 큰 숫자 출력
# 방법 1
if a > b and a> c:
    print(a)
elif b > a and b > c:
    print(b)
else:
    print(c)

# 방법 2
max = a

if max < b:
    max = b
if max < c:
    max = c

print("가장 큰 수",max)

# 가장 작은 수 출력
min = a

if min > b:
    min = b
if min > c:
    min = c

print("가장 작은 수",min)
```

```python
fruits = ['딸기','사과','바나나','배','수박']

# fruits 리스트에 딸기가 포함되어 있다면 딸기가 포함됨 없다면 딸기가 없음
if '딸기' in fruits:
    print("딸기가 포함됨")
else:
    print("딸기가 없음")
```

```python
# [] : 거짓 활용
fruits.clear()

if fruits:
    print("리스트")
else:
    print("빈 리스트")

fruits = ['딸기','사과','바나나','배','수박']
if len(fruits) == 0:
    print("빈 리스트")

fruits.clear()

if fruits:
    print("리스트")
else:
    print("빈 리스트")


if not fruits:
    print("빈 리스트")
```

```python
# and : x and y (x,y 모두 참인 경우 참)
# or : x or y (x,y 둘중 하나만 참인 경우 참)
# not x : x의 True <-> False 바꾸기

# score 가 90이상이고, grade 가 A 인 경우 합격 나머지는 불합격
score, grade = 85, 'A'
if score >= 90 and grade == 'A':
    print("합격")
else:
    print("불합격")

# score 가 90이상이거나 grade 가 A 인 경우 합격 나머지는 불합격
score, grade = 85, 'A'
if score >= 90 or grade == 'A':
    print("합격")
else:
    print("불합격")
```
