````python
# 한 시간 임금은 10700원,
# 8시간 까지는 10700원 으로 지급하고
# 8시간 초과 근무시에는 1.5 배 지급

# 근무시간(hours), rat(시간당 임금), pay(일당)

# 8시간 85600원 + 초과근무 4시간 64,200원
hours,rate,pay = 12, 10700, 0

if hours > 8:
    pay = (rate*8)+((hours-8)*rate*1.5)
    print(pay,'원', sep="")
else:
    pay = (hours*8)*rate
    print(pay,'원', sep="")

print("근무시간 : %d, 일당 : %d" % (hours, pay)) # => %d를 사용해서 정수
print("근무시간 : {}, 일당 : {}".format(hours, pay))
print("근무시간 : {0}, 일당 : {1:.0f}".format(hours, pay))
print(f"근무시간 : {hours}, 일당 : {pay:}")
print(f"근무시간 : {hours}, 일당 : {pay:.0f}")
````

````python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A'] => 리스트 사용
# 1
# B 자동차 입차. 주차장 상태 ==> ['A','B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']

# E 자동차 입차. 주차장 상태 ==> ['A','B','C','D','E']
# 1
# 주차장 만차

# 문자 인코딩 A : 65

parking_lot = []
top, car_name = 0, 'A' # len(parking_lot) => top


while True:
# 사용자는 메뉴 선택
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    if no <= 3:
        if no == 1:
             # top >= 5 일 때 주차장 만차 => 아무것도 안함
            if top >= 5:
                print("주차장 만차")
            # 출력 : A 자동차 입차. 주차장 상태 ==> ['A']
            else:
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name) + 1)
                top = top + 1
        elif no == 2:
            if top > 0:
                outCar = parking_lot.pop()
                print(f"{outCar} 자동차 출차. 주차장 상태 ==> {parking_lot}")
                top = top - 1
                car_name = chr(ord(car_name)-1)
            else:
                print("빠져나갈 자동차가 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해 주세요")
````

````python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A'] => 리스트 사용
# 1
# B 자동차 입차. 주차장 상태 ==> ['A','B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']

# E 자동차 입차. 주차장 상태 ==> ['A','B','C','D','E']
# 1
# 주차장 만차

# 문자 인코딩 A : 65

parking_lot = []
top, car_number = 0, 65 # len(parking_lot) => top

while True:
    num = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    if num == 1:
        # 주차 개수 5개
        # 리스트 안에 순서대로 들어간다 [A B C D E]
        # 리스트에 순서대로 A B C D E 문자를 추가한다
        if top >= 5:
            print(f"만차 주차장 상태 ==> {parking_lot}")
        else:
            parking_lot.append(chr(car_number))
            car_number = car_number + 1
            top = top + 1
        print(f"자동차 입차 : {chr(car_number - 1)} 주차장 상태 ==> {parking_lot}")
    elif num == 2:
        if top == 0:
            print("빠져나갈 자동차 없음")
        else:
            parking_lot.pop()
            car_number = car_number - 1
            top = top - 1
            print(f"자동차 출차 : {chr(car_number)} 주차장 상태 ==> {parking_lot}")
    elif num == 3:
        print("종료")
        break
    else:
        print("잘못된 입력")
        break
````

````python
# 새로운 리스트 생성 : 리스트 컴프리헨션

print([i for i in range(1,7)])
print([i * i for i in range(5)])
print([i * 2 for i in [1,3,7,9,16]])
print([[i, i + 1, i + 2] for i in [1,2,3]])

# 1 ~ 10 사이의 숫자 중에서 홀수 리스트 생성
odd_list = []
for i in range(1,11):
    if i % 2 == 1:
        odd_list.append(i)
print(odd_list)

# 리스트 컴프리헨션
odd_list2 = [i for i in range(1,11) if i % 2 == 1]
print(odd_list2)
````

````python
q_list = ["갑","을","병","정","경"]
# q_list 에서 정을 제외한 요소를 갖는 새로운 리스트 생성
# for
new_list = []
for i in q_list:
    if i != "정":
        new_list.append(i)
print(new_list)
# 리스트 컴프리헨션
w_list = [i for i in q_list if i != "정"]
print(w_list)
# 리스트 컴프리헨션 정답
print([i for i in q_list if i != "정"])
````

````python
q1 = ['nice','study','python','anaconda','!']
# 5글자 이상의 단어만 새로운 리스트로 생성
# len()
print([x for x in q1 if len(x) >= 5])
````

````python
q2 = ['A','b','c','D','e','F','G','h']
# 소문자만 추출 후 새로운 리스트로 생성

# X.islower() => X가 소문자냐 
# X.isupper() => X가 대문자냐 묻는것
# X.isalnum() => X가 알파벳이냐 숫자냐 묻는것
# X.isalpha() => X가 알파벳이냐 묻는것
print([x for x in q2 if x.islower()])
print([x for x in q2 if x.isupper()])
print([x for x in q2 if x.isalpha()])
````
