````python
# 점수가 90 이상이면 A, 80 이상이면 B, 70이상이면 C, 나머지는 F

score = 90

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
````

````python
# and : x and y (x,y 모두 참인 경우 참)
# or : x or y (x,y 둘중 하나만 참인 경우 참)
# not x : x의 True <-> False 바꾸기

# score 가 90이상이고, grade 가 A 인 경우 합격 나머지는 불합격
score, grade = 85, 'A'
if score >= 90 and grade == 'A':
    print("합격")
else:
    print("불합격")

# score 가 90이상이거나 grade 가 A 인 경우 합격 나머지는 불합격
score, grade = 85, 'A'
if score >= 90 or grade == 'A':
    print("합격")
else:
    print("불합격")
````

````python
# 2026년이 윤년 or 평년 인지 판별

year = 2026

# 윤년 조건
# 년도를 4로 나눈 나머지가 0 이고,
# 년도를 100으로 나눈 나머지가 0이 아니거나
# 년도를 400으로 나눈 나머지가 0 이면 윤년

if (year % 4 == 0 and year % 100 != 0) or year % 400 ==0:
    print("윤년")
else:
    print("평년")
````

````python
# 한 시간 임금은 10700원,
# 8시간 까지는 10700원 으로 지급하고
# 8시간 초과 근무시에는 1.5 배 지급

# 근무시간(hours), rat(시간당 임금), pay(일당)

# 8시간 85600원 + 초과근무 4시간 64,200원
hours,rate,pay = 12, 10700, 0

if hours > 8:
    pay = (rate*8)+((hours-8)*rate*1.5)
    print(pay,'원', sep="")
else:
    pay = (hours*8)*rate
    print(pay,'원', sep="")

print("근무시간 : %d, 일당 : %d" % (hours, pay)) # => %d를 사용해서 정수
print("근무시간 : {}, 일당 : {}".format(hours, pay))
print("근무시간 : {0}, 일당 : {1:.0f}".format(hours, pay))
print(f"근무시간 : {hours}, 일당 : {pay:}")
print(f"근무시간 : {hours}, 일당 : {pay:.0f}")
````
