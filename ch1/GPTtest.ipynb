```python
# 문제 1
# 사용자로부터 숫자를 입력받아 짝수이면 "짝수", 홀수이면 "홀수"를 출력하시오.
# 숫자 입력 : 7
num = int(input("숫자 입력"))
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 문제 2
# 사용자로부터 점수를 입력받아 60점 이상이면 "합격", 아니면 "불합격"을 출력하시오.
# 점수 입력 : 85
num = int(input("점수 입력"))
if num >= 60:
    print("합격")
else:
    print("불합격")
```

```python
# 문제 3
# 1부터 10까지 출력하시오.
i = 1
while i <= 10:
    print(i, end=" ")
    i = i + 1
```

```python
# 문제 4
# 1부터 10까지 짝수만 출력하시오.
i = 1
while i <= 10:
    if i % 2 == 0:
        print(i, end=" ")
    i = i + 1
```

```python
# 문제 5
# 1부터 10까지의 합을 구하시오.
i = 1
hap = 0
while i <= 10:
    hap = hap + i
    i = i + 1
print(hap)
```

```python
# 문제 6
# 1부터 20까지 홀수만 출력하시오.
i = 1
while i <= 20:
    if i % 2 != 0:
        print(i,end=" ")
    i = i + 1
```

```python
# 문제 7
# 1부터 100까지의 합을 구하시오.
i = 1
hap = 0
while i <= 100:
    hap = hap + i
    i = i + 1
print(hap)
```
