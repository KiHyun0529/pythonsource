```python
person_list = []
top = 0
person = 'A'

while True:
    num = int(input("[1] 입장 | [2] 퇴장 | [3] 종료"))
    if num == 1:
        if top >= 5:
            print("입장 불가")
        else:
            person_list.append(chr(ord(person)))
            print(f"{person}입장, 현재 ==> {person_list}")
            top = top + 1
            person = chr(ord(person)+1)
    elif num == 2:
        if top > 0:
            person_list.pop()
            person = chr(ord(person)-1)
            print(f"{person}퇴장, 현재 ==> {person_list}")
            top = top - 1
        else:
            print("퇴장 불가")
    elif num == 3:
        print("종료")
        break
    else:
        print("잘못된 입력")
        break
```

```python
car_list = []
top = 0
car_number = 'A'

"[1] 입장 | [2] 퇴장 | [3] 현재 차량 보기 | [4] 종료"

while True:
    num = int(input("[1] 입장 | [2] 퇴장 | [3] 현재 차량 보기 | [4] 종료"))
    if num == 1:
        if top >= 5:
            print("만차")
        else:
            car_list.append(chr(ord(car_number)))
            print(f"{car_number}입차, 주차장 상태 ==> {car_list}")
            top = top + 1
            car_number = chr(ord(car_number) + 1)
    elif num == 2:
        if top > 0:
            car_list.pop()
            car_number = chr(ord(car_number) - 1)
            print(f"{car_number}출차, 주차장 상태 ==> {car_list}")
            top = top - 1
        else:
            print("출차 불가")
    elif num == 3:
        print(f"현재 차량 {car_list}")
    elif num == 4:
        print("종료")
        break
    else:
        print("잘못된 입력")
        break
```
