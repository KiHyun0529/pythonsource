#### 문자열 포매팅

| 코드 | 설명 |
| --- | --- |

| `%x` | 16진수 |
| `%%` | % 문자 자체 |

### 6. 집합 자료형
- 집합에 관련된 것을 쉽게 처리하기 위한 자료형
- 중복 제거 / 순서가 없음 => 인덱싱/슬라이싱 안됨
- 인덱싱, 슬라이싱을 하고 싶다면 가능한 타입으로 변환
- set()

```python
s1 = set()
s2 = set([1,2,3,4,5,4])
s3 = set([1,2,3,4,5,6])
s4 = set([4,5,6,1,2,3])
```

```python
print(s1)
print(s2)
# 중복 제거된 4
print(s3)
print(s4)
# 순서가 없음

#s2[0]
#TypeError: 'set' object is not subscriptable => 순서가 없음
```

```python
# 인덱싱, 슬라이싱을 하고 싶다면 가능한 타입으로 변환
# 리스트 : list()
# 튜플 : tuple()

s1 = "안녕하세요"
print(type(s1)) # str
print(type(3)) # int
print(type(list3)) # list
print(type(t4)) # tuple
print(type(dict2)) # dict
print(type(s2)) # set

# set 자료형 => 새로운 list, tuple (원본그대로)
print(list(s2))
print(tuple(s3))
print(s2)
print(s3)

# 인덱싱 가능
print(list(s2)[2])
print(tuple(s3)[4])
```

```python
# set() : list, tuple 을 set 구조로 변환시 사용 (원본 그대로)
print(list4)
print(set(list4))
print(list4)

print(s1)
print(set(s1))
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1&s2)
print(s1.intersection(s2))

# 합집합
print(s1 | s2)
print(s1.union(s2))

#차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
s1 = "0"
print(type(s1))

# 문자형 => 다른 타입으로 변경

# str => int
print(int(s1))
print(type(int(s1)))
# str => float
print(float(s1))
print(type(float(s1)))
# str => bool
print(bool(s1))
print(type(bool(s1)))
# "0" 문자열이기 때문에 0이여도 True
# 0이 False로 되려면 int로 타입 변경
print(int(s1))
print(bool(int(s1)))
print(type(bool(int(s1))))

s2 = 0
print(bool(s2))
```

```python
# http://naver.com

# 비밀번호 생성
# 규칙 1. http:// 제외 => naver.com
# 규칙 2. 처음 만나는 . 이후 부분은 제외 => naver
# 규칙 3. 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e'개수 + ! 를 조합한 비밀번호 생성
#         => nav51!
# 문자 함수, 슬라이싱, str()

a = 'http://naver.com'

# print(a[7:-4])
b = str(a[7:-4])
print(b) # naver
c = str(b[:3])
print(c) # nav
d = len(b)
print(d)
e = b.count('e')
print(e) # 1
password = str(c) + str(d) + str(e) + '!'
print(password)
```

```python
# 정답

url = 'http://naver.com'

# password = url[8:]
password = url.replace("http://","") # => naver.com
url = url.replace("http://","")
print(url)
# password = password[:5] # => naver
password = password[:password.find(".")] # => naver
# password = password[:3] + len(password) + password.count('e') + '!'
# TypeError: can only concatenate str (not "int") to str
password = password[:3] + str(len(password)) + str(password.count('e')) + '!'
print(password)
```
