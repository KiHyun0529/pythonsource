````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경우
# *
# def 함수명(*변수명):
#     해야할 일

def add(*args):
    result = 0
    for i in args:
        result = result + i
    print(result)
````

````python
add(2,3)
add(2,3,4,5)
add(2,3,4,5,6,7)
````

````python
# def add_mul(choice, *args, a): # choice : 매개변수 / args : 가변 매개변수
#     pass

# 일반 매개변수는 앞 쪽으로 선언
# +, * 실행
def add_mul(choice, a, *args): # choice : 매개변수 / args : 가변 매개변수
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
    return result
````

````python
# add_mul("add", 10, 11, 12, 13)
# add_mul() missing 1 required keyword-only argument: 'a'
# args 가 모든 변수 가져감(10, 11, 12, 13)
print(add_mul("add", 10, 10, 11, 12, 13))
print(add_mul("mul", 10, 10, 11, 12, 13))
# *없는 애들은 지정 해줘야 한다
````

````python
# 키워드 매개변수
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs):
    print(kwargs)

    # key 값만 출력
    for k in kwargs.keys():
        print(k)

    # value 값만 출력
    for v in kwargs.values():
        print(v)

    # key, value 같이 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
# print_kwargs("kim","15")
# TypeError: print_kwargs() takes 0 positional arguments but 2 were given
print_kwargs(name = "kim",age = "15")
print_kwargs(name = "kim",age = "29", addr="서울")
print_kwargs(name = "kim",age = "29", addr="경기", tel="010-1234-5678")
````

````python
# return 옆에 아무것도 쓰지 않기 => 함수 종료

def say_nick(nick):
    if nick == "바보":
        return
    print("나의 별명은 %s 입니다." % nick)
````
